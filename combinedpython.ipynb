```python
# Create a dictionary for each student: {"name": ..., "marks": ...}

# Write a function grade(marks) that returns "A" if marks ≥ 90, "B" if 70–89, "C" if 50–69, else "F".

# Store each student’s grade in the dictionary.

# Save all students to a CSV file grades.csv.

# Skills Tested: Dictionaries, functions, conditionals, loops, CSV I/O.
students={
    "st1": "khair bakhsh",
    "marks":90,
"st2": "ali",
    "marks2":70
}
def grade(marks):
    if marks >=90:
        print("A")
    elif marks<=89 and marks>=70:
        print("B")
    elif marks <=69 and marks>=50:
        print("C")
    else:
        print("F")

for k,v in students.items():
    print(k,v)
grade(students["marks2"])
import csv
with open("store_data.csv","w") as f:
     f=csv.DictWriter(f)
for i in f:
       print(i["st1"],i["marks"])


```
