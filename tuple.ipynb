```python
# Create a tuple of 5 numbers.

# Find the length of a tuple.
tup=(1,2,3,4,5,"b")
print(tup)
```

```python
#tuple slicing
tup=(1,2,3,4,5,"b")
print(tup[1:4])
```

```python
#convert a tuple into a list and add a new element.
s=(1,2,3,4,5)
s=list(s)
print(type(s))
s.append(10)
print(s)
s=tuple(s)
print(type(s))
print(s)
```

```python
#packing of tuples 
pack=1,2,3,4,5
print(pack)
print(type(pack))
```

```python
#unpacking of tuple
unpack=(1,2,3,45,5)
a,b,c,d,e=unpack
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
#swap by unpacking
x=10
y=20
x,y=y,x
print(x,y)
```
