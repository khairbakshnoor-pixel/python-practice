```python
dictionary={
    "name":"khairbakhsh",
    "sport":"cricket",
    "programing":"python and c"
}
dictionary.pop("sport")  
print(dictionary)     # Remove a key
dictionary.popitem()  
print(dictionary)         # Remove the last inserted key-value
del dictionary["name"]  
print(dictionary)        # Delete specific key
dictionary.clear()    
print(dictionary)
```
