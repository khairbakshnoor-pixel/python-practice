```python
class Students:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def avg(self):
        sum=0
        for i in self.marks:
            sum+=i
        print(sum/len(self.marks))
s1=Students("ali",[80,80,80,70])
s1.avg()
```

```python
class calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        print("addition is ",self.a+self.b)
    def sub(self):
        print("sub is ",self.a-self.b)
    def mul(self):
        print("mul  is ",self.a*self.b)
    def div(self):
        print("div is ",self.a/self.b)
        
s=calculator(102,6)
s.add()
s.sub()
s.mul()
s.div()
        
```

```python
class bank:
    def __init__(self,account_holder,balance):
        self.account_holder=account_holder
        self.balance=balance
    def dep(self,amount):
        self.balance+=amount
        print("deposited")
        print(self.show_balance())
    def withdraw(self,amount):
        self.balance-=amount
        print("withdraw")
        print(self.show_balance())
    def show_balance(self):
        return self.balance
s1=bank("babar",10000)
s1.withdraw(100)
s1.dep(1000)
s1.withdraw(900)
s1.show_balance()
```
