```python
#Print a pyramid star pattern.
for i in range(1,5):
    for j in range(5-i):
        print(" ",end="")
    for k in range(2*i-1):
     print("*",end="")
    print()
```

```python
#Print a repyramid star pattern.
for i in range(5,0,-1):
    for j in range(5-i):
        print(" ",end="")
    for k in range(2*i-1):
     print("*",end="")
    print()
```
