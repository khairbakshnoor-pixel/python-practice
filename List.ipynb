```python
# Find the largest and smallest element in a list.
listt=[2,4,5,6,7,8,111,4,]
largest=listt[0]
smallest=listt[0]
for i  in listt:
    if i>largest:
        largest=i
print(largest)
for i  in listt:
    if i<smallest:
        smallest=i
print(smallest)

```

```python
# Count how many times a specific number appears in a list.
lis=[1,2,3,5,4,3,4,5]
num=3
count=0
for i in lis:
    if i ==num:
        count=count+1
print(f'the number {num}comes {count } times')
 

```

```python
# 2. Input list, check for each element weather 
# is it positive or negative, store &
# display both in different lists
numbers=[]
positive=[]
negative=[]
n=int(input("Enter number of elements in list"))
for i in range(n):
    element=int(input("Enter the element"))
    numbers.append(element)

for num in numbers:
    if num > 0:
        positive.append(num)
    elif num < 0:
        negative.append(num)

print("Positive numbers:", positive)
print("Negative numbers:", negative)
```
