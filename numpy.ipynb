```python
#vectorized math functions
import numpy as np
array=np.array([1,2,3,4])
print(np.sqrt(array))
print(np.round(array))
print(np.ceil(array))
print(np.floor(array))
print(np.amax(array))
print(np.amin(array))
print(np.absolute(array))
print(np.average(array))
print(np.std(array))
print(np.mean(array))
print(np.sum(array))
print(np.argmin(array))
print(np.argmax(array))
print(np.sum(array,axis=0))
```
