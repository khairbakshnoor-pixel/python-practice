```python
# Factorial Function

# Write a function factorial(n) to compute the factorial of a number using a loop.
def fact(n):
    facti=1
    
    for i in range(1,n+1):
        facti=facti*i
    return facti

n=int(input("enter a number"))
print(fact(n))
```

```python
# Sum of List

# Write a function sum_list(lst) that returns the sum of all elements in a list.
def sum(n):
    sum=0
    for i in range(0,5):
        sum+=n[i]
    return sum
mylist=[2,3,4,5,10]
print("the sum of the list is ",sum(mylist))
```
